```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Assuming 'merged' is your DataFrame containing the 'Clients_IFSSA.preferred_languages' column

# STEP 1: Split the 'preferred_languages' column by a delimiter
merged['preferred_languages_split'] = merged['Clients_IFSSA.preferred_languages'].str.split(',')

# STEP 2: Flatten the lists and clean up any leading/trailing spaces
all_languages = [lang.strip() for sublist in merged['preferred_languages_split'].dropna() for lang in sublist]

# STEP 3: Count the occurrences of each language
language_counts = Counter(all_languages)

# STEP 4: Convert to a DataFrame to see the counts more easily
language_counts_df = pd.DataFrame(language_counts.items(), columns=['Language', 'Count'])

# Sort the DataFrame by 'Count' in descending order and keep only the top 10
language_counts_df = language_counts_df.sort_values(by='Count', ascending=False).head(10)

# Print the sorted DataFrame
print(language_counts_df)

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(language_counts_df['Language'], language_counts_df['Count'], color='skyblue')
plt.title('Top 10 Language Preferences Count')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x labels for better visibility

# Add count labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

## **Fill**

```python
# list of columns to fill missing values for
columns_to_fill = [
    "Clients_IFSSA.sex_new",
    "Clients_IFSSA.address",
    "Clients_IFSSA.age",
    "Clients_IFSSA.dependents_qty",
    "Clients_IFSSA.household",
    "Clients_IFSSA.latest_language_code"
]

# defining a function to fill NaN values with the mode for specified columns
def fill_nulls_with_mode(df, group_col, fill_cols):
    for col in fill_cols:
        # firstly calculating the mode for each group
        mode_series = df.groupby(group_col)[col].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
        # then filling in the null values into the original df
        df[col] = df[col].fillna(df[group_col].map(mode_series))
    return df

# filling in the null values by calling the "fill_nulls_with_mode" function
merged = fill_nulls_with_mode(merged, "Clients_IFSSA.unique id", columns_to_fill)
```

## Summary of Handling Null Values in the Dataset

We attempted to fill the null values in the following columns of the dataset using the Clients_IFSSA.unique id column to ensure that each client’s information was accurately represented:

**Columns with Null Values:**

However, there were no changes in the count of null values after this process. This was likely because for many of the client IDs with missing values, the corresponding rows also had all other fields null. As a result, there were no mode values available to fill the nulls, leading to the persistence of null values in the dataset.

```python
# importing the datatime library
from datetime import datetime

# creating a function to calculate age from given birthdate
def calculate_age(birthdate):
    # firstly checking if the birthdate is not null
    if pd.notnull(birthdate):
        # then getting the current date and time
        today = datetime.now()
        # finally calculatig the age by subtracting (-) birth year from current year & adjusting if today's date is before the birthday
        return today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    # returning 'None' if the birthdate is null
    return None

# STEP 1: converting 'primary_birthdate' to datetime if it's not already
merged['Clients_IFSSA.primary_birthdate'] = pd.to_datetime(merged['Clients_IFSSA.primary_birthdate'], errors='coerce')

# STEP 2: grouping by 'Clients_IFSSA.unique id' and filling in 'Clients_IFSSA.age' using 'primary_birthdate'
for unique_id in merged['Clients_IFSSA.unique id'].unique():
    # getting the subset for the current unique_id
    subset = merged[merged['Clients_IFSSA.unique id'] == unique_id]

    # using a if-statement = if 'Clients_IFSSA.primary_birthdate' is not null, then we calculate age and fill 'Clients_IFSSA.age'
    if not subset['Clients_IFSSA.primary_birthdate'].isnull().all():
        # taking the first non-null birthdate
        birthdate = subset['Clients_IFSSA.primary_birthdate'].dropna().iloc[0]
        age = calculate_age(birthdate)

        # filling the age where it's null
        merged.loc[merged['Clients_IFSSA.unique id'] == unique_id, 'Clients_IFSSA.age'] = merged.loc[merged['Clients_IFSSA.unique id'] == unique_id, 'Clients_IFSSA.age'].fillna(age)
```

```python
# defining Ramadan date ranges for both years first
ramadan_2023 = (pd.Timestamp('2023-03-22'), pd.Timestamp('2023-04-21'))
ramadan_2024 = (pd.Timestamp('2024-03-11'), pd.Timestamp('2024-04-09'))

# creating a new column in the original df to identify Ramadan dates
# firstly defining a function which takes "row" as input
def identify_ramadan_dates(row):
    # using if statement to iterate through each date in both ramada variabls
    # checking to see if the pickup_date falls within Ramadan 2023 or 2024, otherwise label it as 'Other'
    if ramadan_2023[0] <= row['pickup_date'] <= ramadan_2023[1]:
        return 'Ramadan 2023'
    elif ramadan_2024[0] <= row['pickup_date'] <= ramadan_2024[1]:
        return 'Ramadan 2024'
    else:
        return 'Other'

# applyig the "identify_ramadan_dates" function to each row to create the 'Ramadan_Period' column in the df
merged_data['Ramadan_Period'] = merged_data.apply(identify_ramadan_dates, axis=1)

# grouping by 'Ramadan_Period' and counting the number of pickups (rows)
# 'total_pickups' is a newly created column that counts the number of pickups for each Ramadan_Period
# this is to compare total pickups for Ramadan vs. other periods
monthly_pickups = merged_data.groupby('Ramadan_Period').size().reset_index(name='total_pickups')

# visulizing
plt.figure(figsize=(8, 5))
plt.bar(monthly_pickups['Ramadan_Period'], monthly_pickups['total_pickups'], color=['pink', 'blue', 'orange'])
plt.xlabel('Period')
plt.ylabel('Total Hamper Pick-ups')
plt.title('Hamper Pick-ups During Ramadan (2024) vs Other')
plt.show()
```

For this feature, we will calculate the time difference between the first visit and the current visit for each client.

#**Block 5: Understanding Feature Relationships (Correlation Analysis & Visualization) (%20):**

Before jumping into the model, it's essential to analyze the relationships between features to get a sense of how they interact with each other and the target variable. Visualizations such as correlation matrices and scatterplots help identify which features are more influential.


**What is Correlation?**
- Correlation measures the linear relationship between two variables. A positive correlation means that as one variable increases, the other tends to increase. A negative correlation means that as one increases, the other tends to decrease.
- Correlation matrix provides a tabular representation of how strongly features are related to each other. A heatmap makes it easy to visualize these correlations.


**Steps to Perform Correlation Analysis:**
1. Correlation Matrix: Compute the correlation matrix between the features and the target variable.
2. Heatmap: Visualize the correlation matrix using a heatmap to easily spot highly correlated or non-correlated features.
3. Scatter Plots: Use scatter plots to visually assess the relationships between key features and the target.
4. Feature Selection: Based on the correlations, make decisions about which features to keep, combine, or drop.

###***Please consider the visulizations from block 2 of feature engineering as part of this section as well***

```python
correlation_matrix = merged_data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix')
plt.show()

# Choose specific features for scatter plots (example: 'total_quantity_collected' vs. 'total_dependents_last_6_months')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_data, x='total_quantity_collected', y='total_dependents_last_6_months')
plt.title('Scatter Plot: Total Quantity Collected vs. Total Dependents Last 6 Months')
plt.xlabel('Total Quantity Collected')
plt.ylabel('Total Dependents Last 6 Months')
plt.show()
```
