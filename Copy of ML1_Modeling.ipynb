# **Block 2: Model Selection & Training**


```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Define the target variable
target = time_lagged_features['pickup_date_count']

# Define the features (excluding the target variable and date)
features = time_lagged_features[['scheduled_date_count',
                                  'pickup_date_count_lag_7',
                                  'scheduled_date_count_lag_7',
                                  'pickup_date_count_lag_14',
                                  'scheduled_date_count_lag_14',
                                  'pickup_date_count_lag_21',
                                  'scheduled_date_count_lag_21']]

# Data Splitting using Time Series Cross-Validation
tscv = TimeSeriesSplit(n_splits=5)  # Adjust the number of splits as necessary
mae_scores_arima = []  # List to store evaluation metrics for ARIMA models
mae_scores_rf = []     # List for Random Forest model
mae_scores_xgb = []    # List for XGBoost model

# Cross-validation for ARIMA model evaluation
for train_index, test_index in tscv.split(target):
    # Split the data into training and testing sets
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]

    # Fit the ARIMA model (example: ARIMA(1, 1, 1))
    model_arima = ARIMA(y_train, order=(1, 1, 1))
    model_arima_fit = model_arima.fit()

    # Forecast for the length of the test set
    y_pred_arima = model_arima_fit.forecast(steps=len(y_test))

    # Evaluate ARIMA model performance using Mean Absolute Error
    mae_arima = mean_absolute_error(y_test, y_pred_arima)
    mae_scores_arima.append(mae_arima)

    # Print the results for this split
    print(f"ARIMA - Train indices: {train_index}, Test indices: {test_index}, MAE: {mae_arima}")

# Normalize features for machine learning models
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Cross-validation for Random Forest model
for train_index, test_index in tscv.split(target):
    # Split the data into training and testing sets
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]
    X_train, X_test = features_scaled[train_index], features_scaled[test_index]

    # Fit the Random Forest model
    model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
    model_rf.fit(X_train, y_train)

    # Make predictions
    y_pred_rf = model_rf.predict(X_test)

    # Evaluate Random Forest model performance
    mae_rf = mean_absolute_error(y_test, y_pred_rf)
    mae_scores_rf.append(mae_rf)

    # Print the results for this split
    print(f"Random Forest - Train indices: {train_index}, Test indices: {test_index}, MAE: {mae_rf}")

# Cross-validation for XGBoost model
for train_index, test_index in tscv.split(target):
    # Split the data into training and testing sets
    y_train, y_test = target.iloc[train_index], target.iloc[test_index]
    X_train, X_test = features_scaled[train_index], features_scaled[test_index]

    # Fit the XGBoost model
    model_xgb = XGBRegressor(n_estimators=100, random_state=42)
    model_xgb.fit(X_train, y_train)

    # Make predictions
    y_pred_xgb = model_xgb.predict(X_test)

    # Evaluate XGBoost model performance
    mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
    mae_scores_xgb.append(mae_xgb)

    # Print the results for this split
    print(f"XGBoost - Train indices: {train_index}, Test indices: {test_index}, MAE: {mae_xgb}")

# Print the average MAE for all models
average_mae_arima = sum(mae_scores_arima) / len(mae_scores_arima) if mae_scores_arima else 0
average_mae_rf = sum(mae_scores_rf) / len(mae_scores_rf) if mae_scores_rf else 0
average_mae_xgb = sum(mae_scores_xgb) / len(mae_scores_xgb) if mae_scores_xgb else 0

print(f"\nAverage MAE for ARIMA: {average_mae_arima}")
print(f"Average MAE for Random Forest: {average_mae_rf}")
print(f"Average MAE for XGBoost: {average_mae_xgb}")

# ACF and PACF plots for ARIMA parameter tuning
plot_acf(target)
plt.title('ACF Plot')
plt.show()

plot_pacf(target)
plt.title('PACF Plot')
plt.show()
```

# Block 3: Model Evaluation & Diagnostics (%25)

```python
# Import necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Function to calculate performance metrics
def calculate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mae, mse, rmse, mape

# Plotting predictions vs. actual values
def plot_predictions(y_test, y_pred, model_name):
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.index, y_test, label='Actual', color='blue')
    plt.plot(y_test.index, y_pred, label='Predicted', color='red', linestyle='--')
    plt.title(f'{model_name} Predictions vs Actual Values')
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.legend()
    plt.show()

# Evaluating ARIMA Model
mae_arima, mse_arima, rmse_arima, mape_arima = calculate_metrics(y_test, y_pred_arima)
print(f"ARIMA Model Metrics:\nMAE: {mae_arima}\nMSE: {mse_arima}\nRMSE: {rmse_arima}\nMAPE: {mape_arima}%")
plot_predictions(y_test, y_pred_arima, "ARIMA")

# Evaluating Random Forest Model
mae_rf, mse_rf, rmse_rf, mape_rf = calculate_metrics(y_test, y_pred_rf)
print(f"Random Forest Model Metrics:\nMAE: {mae_rf}\nMSE: {mse_rf}\nRMSE: {rmse_rf}\nMAPE: {mape_rf}%")
plot_predictions(y_test, y_pred_rf, "Random Forest")

# Evaluating XGBoost Model
mae_xgb, mse_xgb, rmse_xgb, mape_xgb = calculate_metrics(y_test, y_pred_xgb)
print(f"XGBoost Model Metrics:\nMAE: {mae_xgb}\nMSE: {mse_xgb}\nRMSE: {rmse_xgb}\nMAPE: {mape_xgb}%")
plot_predictions(y_test, y_pred_xgb, "XGBoost")

# Visualization of all metrics
metrics_df = pd.DataFrame({
    'Model': ['ARIMA', 'Random Forest', 'XGBoost'],
    'MAE': [mae_arima, mae_rf, mae_xgb],
    'MSE': [mse_arima, mse_rf, mse_xgb],
    'RMSE': [rmse_arima, rmse_rf, rmse_xgb],
    'MAPE (%)': [mape_arima, mape_rf, mape_xgb]
})

plt.figure(figsize=(12, 6))
metrics_df.set_index('Model').plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title('Model Evaluation Metrics')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
